# NUMPY (Numerical Python)

Permet de manipuler des array = liste Python dont les éléments sont tous du même type, ce qui permet d'optimiser le stockage en mémoire et donc la vitesse d'éxécution.

Si dans la liste de départ il y a des données de types différents, NumPy essaiera de tous les convertir au type le plus général.

Tableau 1D

```python
import numpy as np
liste = [1, 3.0, 5, 8, 9, 10 ,2]
array = np.array(liste)
print(type(liste))
print(type(array))
print(array.dtype) # retourne le type des éléments contenu dans l'array
```

Comme en Matlab, il est possible d'utiliser la syntaxe : nom_array[i:j:p] , avec :

* i : l’indice du début ;
* j : l’indice de fin ;
* p : le pas.

```python
print("l'élément 0 : ", array[0])
print("l'élément 1 : ",array[1])
print("le premier en partant de la fin : ",array[-1])
print("le deuxième en partant de la fin : ",array[-2])
print("les éléments 0 à 3 : ",array[0:3])
print("les éléments 0 à 6 par pas de 2 : ",array[0:3:6]) # si début vide : 0, si pas vide : 1, fin vide : la fin
print("les 2 premiers : ",array[:2])
print("après les 2 premiers : ",array[2:])
print("du 4e en partant de la fin au 2e-1 en partant de la fin : ",array[-4:-2])
print("les éléments qui sont supérieur à 5 : ", array[array>5], "car :", array>5)
print("Et (pas de and ici comparaison car terme à terme) : ", array[(array>5) & (array<9)])
print("Ou (pas de or ici comparaison car terme à terme) : ", array[(array>5) | (array<9)])
```

```python
 # début vide = 0, pas vide = 1, fin = indice final => on affiche tout
 # comme print(array), inutile mais à savoir pour comprendre l'indiçage
print(array[::])

array.sort()
print(array) # le tableau est modifié attention !
```

## Quelques Fonctions

```python
# calculer la moyenne
print(array.mean())

# calculer le maximum (ou le minimum)
print(array.max())
print(array.min())

# accéder à l’indice de l’élement minimum (ou maximum)
print(array.argmin())
print(array.argmax())

# ordonner par ordre croissant => array est modifié
array.sort()
print(array)

# calculer la somme
print(array.sum())

# Matlab like
i = 0
j = 50
n = 5
m = 2
print(np.linspace(i, j, n)) # [0,50] with linear space
print(np.arange(i, j, n)) # [0,50] with step 5
print( np.ones((n,m)) ) # /!\ c'est un tuple à l'intérieur
print( np.zeros((n,m)) )

# un tableau de 6x3 rempli de valeurs aléatoires comprises entre 0 et 1
print( np.random.random((6, 3)) )

# un tableau de 3x3 rempli de valeurs aléatoires entières, comprises entre 1 et 10
print( np.random.randint(1, 10, size=(3, 3)) )
```

## Calcul matriciel

```python
A = np.array([[1, 2], [3, 4]])
B = np.array([[5, 10], [15, 20]])

print(A+B) # addition terme a terme
print(A-B) # soustraction terme a terme
print(A*B) # produit terme a terme
print(A**B) # puissance terme a terme

# produit matriciel, 2 ecritures
print( np.dot(A,B) )
print( A@B ) # seulement pour v.recent de Python
```

## Selection par index ou condition 'loc'

- mon_dataframe.loc[ condition sur les lignes, colonne(s) ]
- mon_dataframe.loc[ index, numéro de colonne ]

```python
data.loc[data['genre'] == 'M', :]
```

```python
display(data.loc[0:10, :])
display(data.loc[0:10, :].index)
display(data.iloc[0:10, :])
print(data.iloc[0:10, :].index)
```

```python
# création d’un data frame trié avant de ‘désordonner’ les Index
df_temp = data.sort_values('email')
display(df_temp)
display(df_temp.reset_index())
display(df_temp.reset_index(drop=True)) # sans garder les anciens Index
```

L'exemple ci-dessous montre la différence entre indices et index :
- L' **indice** est la position dans le data frame.
- L' **index** est généré à l'init et identifie un élément.

Par conséquent, déplacer le 133 élément à la position 0 revient à modifier son indice de 132 à 0 alors que son index reste 132, à moins d'en forcer la modification comme indiqué au dessus.

La méthode **loc** permet de selectionner par **index** alors que **iloc** permet la selection par **indice**
